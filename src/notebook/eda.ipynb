```python
def mostrar_precentiler_outliers(dataframe, columna):
    percentiles = dataframe[columna].quantile([0.05, 0.95])
    valores = dataframe[columna]
    outliers_percentiles = valores[(valores < percentiles[0.05]) | (valores > percentiles[0.95])]

    # Visualiza los outliers
    plt.hist(valores, color='g')
    plt.scatter(outliers_percentiles, np.zeros_like(outliers_percentiles), color='r')
    plt.title(f"Outliers de {columna}")
    plt.show()
```
